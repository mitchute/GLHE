```python
from scipy.interpolate import interp1d
from scipy.misc import derivative

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [15, 9]
plt.rcParams['font.size'] = 14

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python
data = np.genfromtxt("g_functions.csv", delimiter=',')
```

```python
f = interp1d(data[:, 0], data[:, 1], fill_value='extrapolate')
```

```python
x = np.arange(data[0, 0], data[-1, 0], step=0.01)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(data[:, 0], data[:, 1], linestyle='', marker='D', label="Data")
ax.plot(x, f(x), label='f(x)')
ax.plot(x, derivative(f, x), label="f'(x)")
ax.plot(x, derivative(interp1d(x, derivative(f, x), fill_value='extrapolate'), x), label='f"(x)')

plt.legend()
plt.show()
```

```python
lntts = np.arange(-14.5, 3, step=0.1)
lntts_2 = np.arange(-14.5, 3, step=0.2)
L = 76.2
k = 2.7
rho = 2500
cp = 880
alpha = k/(rho * cp)
ts = L ** 2 / (9 * alpha)
```

```python
t_ts = np.exp(lntts)
t_ts_2 = np.exp(lntts_2)

t = ts * t_ts
t_2 = ts * t_ts_2
```

```python
plt.semilogy(t)
plt.semilogy(t_2)
```

```python
d_lntts = 1 / t_ts
d_lntts_2 = 1 / t_ts_2
```

```python
plt.semilogy(d_lntts)
plt.semilogy(d_lntts_2)
```

```python
plt.semilogy(lntts, d_lntts)
plt.semilogy(lntts_2, d_lntts_2)
```
